## Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
plt.figure(figsize=(6,4), dpi=100)
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# same figure in matplotlib!
plt.figure(figsize=(6,4), dpi=100)
plt.scatter(x='total_bill', y='tip', data=tips)
```

## Distribution Plots

### rugplot

```python
tips.head()
```

```python
sns.rugplot(x=tips['total_bill'])
```

```python
sns.rugplot(x='total_bill', data=tips)
```

```python
sns.rugplot(x='total_bill', data=tips, height=0.5)
```

### displot & histplot

```python
sns.displot(x='total_bill', data=tips)
```

```python
sns.histplot(x='total_bill', data=tips)
```

```python
sns.displot(x='total_bill', data=tips, col='day')
```

```python
sns.histplot(x='total_bill', data=tips, hue='day')
```

```python
sns.histplot(x='total_bill', data=tips, kde=True)
```

```python
sns.histplot(x='total_bill', data=tips, kde=True, bins=20)
```

```python
plt.figure(figsize=(12,8))
sns.set(style='whitegrid')
sns.histplot(x='total_bill', data=tips, bins=20, kde=True)
```

### kdeplot

```python
sns.kdeplot(x='total_bill', data=tips)
```

```python
sns.kdeplot(x='total_bill', data=tips, clip=[5,30])
```

```python
sns.kdeplot(x='total_bill', y='tip', data=tips)
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(x='total_bill', y='tip', data=tips)
sns.scatterplot(x='total_bill', y='tip', data=tips, color='r')
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(x='total_bill', y='tip', data=tips)
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

### scatterplot

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time', size='smoker')
```

```python
plt.figure(figsize=(6,4))
fig,ax=plt.subplots(1,2)
sns.scatterplot(data=tips, x='total_bill', y='tip',ax=ax[0]);
sns.kdeplot(data=tips, x='total_bill', y='tip', color='red', shade=True,ax=ax[1])
```

```python
fig,ax=plt.subplots(1,4,figsize=(18,8))
sns.scatterplot(x='total_bill', y='tip',data=tips, ax=ax[0])
sns.lineplot(x='total_bill', y='tip',data=tips, ax=ax[1])
sns.regplot(x='total_bill', y='tip',data=tips, ax=ax[2])
sns.kdeplot(x='total_bill', y='tip',data=tips, ax=ax[3])
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind= 'line', col='day')
```

```python
sns.kdeplot(data=tips, x='total_bill', y='tip', color='brown', shade=True)
```
