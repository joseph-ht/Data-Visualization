## 2: Plot 12 most dangerous cities

```python
df.city.value_counts().head(12)
```

```python
most_dangerous = pd.DataFrame(df.city.value_counts().head(12))
most_dangerous.head()
```

```python
fig=plt.figure(figsize=(10,6))
a = sns.barplot(y=most_dangerous.city, x=most_dangerous.index)
a.set_xticklabels(a.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
city = df.city.value_counts().head(12).index
city
```

```python
fig=plt.figure(figsize=(10,6))
b = sns.countplot(x=df.city, order=df.city.value_counts().head(12).index)
b.set_xticklabels(b.get_xticklabels(), rotation=45, horizontalalignment='right')
b.set_title('Most Dangerous Cities', color='blue', fontsize=15)
plt.show()
```

## 3: Plot 20 most dangerous states

```python
df.head()
```

```python
df.state.value_counts().iloc[:30]
```

```python
fig=plt.figure(figsize=(10,6))
sns.barplot(x=df.state.value_counts().iloc[:30].index, y=df.state.value_counts().iloc[:30].values,data=df)
plt.title('Most Dangerous States', color='blue', fontsize=15);
```

```python
fig=plt.figure(figsize=(10,6))
c = sns.countplot(x=df.state, order=df.state.value_counts().iloc[:30].index)
c.set_xticklabels(c.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('Most Dangerous States', color='blue', fontsize=15);
```
