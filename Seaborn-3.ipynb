# Seaborn - 3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

## Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # kind=scatter default
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', shade=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='red')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', hue='sex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='orange')
```

```python
sns.jointplot(x = 'total_bill', y='tip', data=tips, height=8)
```

```python
sns.scatterplot(x='total_bill' ,y='tip', data=tips)
sns.kdeplot(x='total_bill' ,y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

## Catplot

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips);
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='strip')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='strip', jitter=False)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', col='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box', col='sex', hue='time')
```

```python
sns.relplot(x='day', y='total_bill', data=tips, kind='line', col='sex', hue='time')
sns.relplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', hue = 'time', kind = 'line')
```

## Pairplot

```python
g=sns.PairGrid(tips)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(tips)
g.map_upper(sns.scatterplot)
```

```python
g=sns.PairGrid(tips)
g.map_upper(sns.kdeplot)
```

```python
g=sns.PairGrid(tips)
g.map_lower(sns.scatterplot)
```

```python
g.map_diag(sns.histplot)
```

```python
g.map_lower(sns.kdeplot, color='red')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot, color='orange')
```

```python
g=sns.PairGrid(tips, corner=True)
g.map_lower(sns.kdeplot, color='red')
g.map_diag(sns.histplot)
```

## FacetGrid

```python
sns.FacetGrid(tips)
```

```python
g=sns.FacetGrid(data = tips, row='smoker', col='time')
```

```python
g=sns.FacetGrid(data = tips, row='smoker', col='time')
g.map(plt.hist, 'total_bill')
```

```python
g=sns.FacetGrid(data = tips, row='smoker', col='time')
g.map(plt.scatter, 'total_bill', 'tip')
```

```python
g=sns.FacetGrid(data = tips, row='smoker', col='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
```
