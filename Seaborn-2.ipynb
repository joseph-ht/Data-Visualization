```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
tips.info()
```

### Countplot

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x = 'day', data=tips)
plt.show()
```

```python
sns.countplot(x='day', data=tips, hue='sex')
plt.legend(loc=2)
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.info()
```

```python
sns.countplot(x='cylinders', data=mpg);
```

```python
sns.set_style('white')
```

```python
sns.countplot(x='model_year', data=mpg );
```

```python
sns.countplot(y='model_year', data=mpg );
```

### Stripplot

```python
sns.stripplot(x=tips['total_bill']);
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=tips['day'], y=tips['total_bill'])
```

```python
sns.pairplot(tips)
```

scatter plot is for the continious variables, stripplot is for the dicrete variables !

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=tips['size'], y=tips['total_bill'], hue=tips['sex'])
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=tips['size'], y=tips['total_bill'], hue=tips['sex'], dodge=True) #dodge groupes hue variables
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x=tips['size'], y=tips['total_bill'], hue=tips['sex'], jitter=False)
```
