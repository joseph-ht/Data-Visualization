# Functional Method

```python
fig = plt.figure()

ax1 = fig.add_axes([0.0, 0.0, 1, 1])
ax2 = fig.add_axes([1.1, 0.3, 0.4, 0.4])

ax1.plot(age, salary, 'r')
ax1.set_xlabel('age')
ax1.set_ylabel('salary')
ax1.set_title('Salary over age')

ax2.plot(age, salary_2, 'y')
ax2.set_xlabel('age')
ax2.set_ylabel('salary_2')
ax2.set_title('salary_2 over age')

plt.show()
```

# Subplot function

```python
plt.subplot(2, 1, 1)
plt.plot(age, salary, 'r')

plt.subplot(2, 1, 2)
plt.plot(age, salary_2, 'b')

plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(age, salary, 'r')
ax[1].plot(age, salary_2, 'b')

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(age, salary, 'r')

plt.subplot(1, 2, 2)
plt.plot(age, salary_2, 'b')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(age, salary, color='r', linestyle='--', marker='s')

plt.subplot(2, 2, 2)
plt.plot(age, salary_2, 'b', ls='-.', marker='D')

plt.subplot(2, 2, 3)
plt.plot(age, salary, 'g', marker='H')

plt.subplot(2, 2, 4)
plt.plot(age, salary_2, 'y', marker='X', ls=':')

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(2, 3)
ax[0][0]
ax[1][2].plot(age, salary)

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 6))

ax[0].plot(age, salary, 'r')
ax[1].plot(age, salary_2, 'b')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle('Different line widths and styles')

ax.plot(x, x-1, color='r', linewidth=0.25)
ax.plot(x, x-2, color='r', linewidth=0.50)
ax.plot(x, x-3, color='r', linewidth=0.75)
ax.plot(x, x-4, color='r', linewidth=10);
```

```python
plt.plot(age, salary, label='salary 1')
plt.plot(age, salary_2, label='salary 2')
plt.xlabel('age')
plt.ylabel('salary')
plt.title('salary over age')

plt.legend(loc=4)

plt.show()
```

```python
x = np.arange(1, 10)

fig, ax = plt.subplots(1, 2, figsize=(10, 6))

ax[0].plot(x, x**2, x, x**3, x, x**4, 'r')
ax[1].plot(x, x**2, x, x**3, x, x**4, 'r')

ax[1].set_xlim([1, 3])
ax[1].set_ylim([1, 10])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(tips.groupby('day')['total_bill'].sum().index, tips.groupby('day')['total_bill'].sum().values, label='total bill')
ax.bar(tips.groupby('day')['total_bill'].count().index, tips.groupby('day')['total_bill'].count().values, label='tips')
plt.legend()

plt.show()
```
